**1.WSGI（Web Server Gateway Interface）服务器网管接口**  
Python应用程序和Web服务器间的一种接口。就像一种协议，只要遵循协议，WSGI应用就能在任何服务器运行。

**2.uWSGI**  
就像一个应用程序，能帮助实现WSGI协议，Http协议。

**3.代理服务器**  
允许一个终端，通过服务器与另一个终端进行非直接连接。这能保障该终端的隐私和安全，隐藏其IP。如VPN服务。

**4.反向代理服务器**  
代理服务器接收客户端的请求，将请求转发到内网处理，然后返回Response给客户端。这保证了内网安全，同时可以优化网站负载，它还能缓存静态资源，加快访问速度。
